```python
import mysql.connector

conn = mysql.connector.connect(
    host="studentdb.crsqkeh7g4o2.us-east-2.rds.amazonaws.com",
    user="living_wage",
    password="LivingWage123!",
    database="giq_data",
    port=3306,
    charset="utf8mb4",
    collation="utf8mb4_unicode_ci"
)

cursor = conn.cursor()
cursor.execute("SHOW TABLES;")
tables = cursor.fetchall()

for (table_name,) in tables:
    print(f"\nTable: {table_name}")
    
    # Get column names
    cursor.execute(f"DESCRIBE {table_name};")
    columns = [col[0] for col in cursor.fetchall()]
    
    # Build query to count NULLs
    null_checks = ", ".join(
        [f"SUM(CASE WHEN `{col}` IS NULL THEN 1 ELSE 0 END) AS `{col}_nulls`" for col in columns]
    )
    query = f"SELECT {null_checks} FROM `{table_name}`;"
    
    cursor.execute(query)
    null_counts = cursor.fetchone()
    
    # Print nicely
    for col, null_count in zip(columns, null_counts):
        print(f"{col}: {null_count} nulls")

cursor.close()
conn.close()
```

```python
def normalize_location_col(df):
    for col in df.columns:
        if col.lower() == "location_codes" or col.lower() == "location_code":
            if col != "location_codes":
                df = df.rename(columns={col: "location_codes"})
            return df
    raise ValueError("cann't find location_codes column")

compensation_df = normalize_location_col(compensation_df)
expenditure_df = normalize_location_col(expenditure_df)
gs_prices_df = normalize_location_col(gs_prices_df)
housing_df = normalize_location_col(housing_df)
```

```python
# Convert to USD
def add_usd_column(df, value_cols, rate_col="rate"):
    for col in value_cols:
        if col in df.columns:
            df[f"{col}_usd"] = df[col] / df[rate_col]
    return df

compensation_df = add_usd_column(compensation_df, ["base_pay"])
housing_df = add_usd_column(housing_df, ["price", "utilities"])
gs_prices_df = add_usd_column(gs_prices_df, ["Low_Price"])
```
