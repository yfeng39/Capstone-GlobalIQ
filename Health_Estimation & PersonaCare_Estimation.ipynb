```python
health_df = gs_prices_df[gs_prices_df["Category"].eq("Health")].copy()
```

```python
# Assign annual frequency weights (from WHO / OECD sources)
freq_map = {
    'Painkiller': 6,                # OECD/CDC: common OTC analgesic use ~6x/yr
    'Medical Doctor': 5,            # OECD avg doctor visits/person ~5x/yr
    'Cough Drops': 4,               # CDC: 24 colds/yr
    'Contact Lens Solution': 6,     # regular users ~monthly refill
    'Band Aids/Plasters': 2,        # Euromonitor: wound care ~2x/yr
    'Isopropyl Alcohol': 1,         # disinfectant ~1x/yr
    'Vitamins': 12,                 # supplements often monthly
    'Penicillin': 1,                # antibiotics ~1x/yr
    'Dental cleaning & checkup': 2  # dentist visits ~2x/yr
}
```

```python
health_df["freq_weight"] = health_df["Final_Item"].map(freq_map)
```

```python
import numpy as np
# Compute weighted pharmacy/health index per country
def weighted_mean(x):
    return np.average(x["Low_Price_usd"], weights=x["freq_weight"])

pharmacy_weighted = (
    health_df
    .groupby("Countries", as_index=False)
    .apply(lambda g: pd.Series({"weighted_pharmacy_index": weighted_mean(g)}),
           include_groups=False)
)
```

```python
# Fetch World Bank OOP expenditure data (indicator SH.XPD.OOPC.CH.ZS)
import requests
import io
import zipfile

url = "https://api.worldbank.org/v2/en/indicator/SH.XPD.OOPC.CH.ZS?downloadformat=csv"
r = requests.get(url)
r.raise_for_status()

zip_file = io.BytesIO(r.content)
with zipfile.ZipFile(zip_file) as z:
    data_files = [f for f in z.namelist() if f.startswith("API_") and f.endswith(".csv")]
    print("All CSV files in ZIP:", data_files)
    if not data_files:
        raise FileNotFoundError(" No valid data file found in World Bank ZIP.")
    
    data_file = data_files[0]
    print(f"Reading data from: {data_file}")
    
    oop_data = pd.read_csv(z.open(data_file), skiprows=4)

print("Columns in oop_data:", oop_data.columns.tolist())
print(oop_data.head())
```

```python
# Clean & reshape
oop_data = oop_data.drop(columns=[col for col in oop_data.columns if "Unnamed" in col], errors="ignore")

oop_df = oop_data.melt(
    id_vars=["Country Name", "Country Code"],
    value_name="oop_share",
    var_name="year"
)

oop_df["year"] = pd.to_numeric(oop_df["year"], errors="coerce")
oop_df["oop_share"] = pd.to_numeric(oop_df["oop_share"], errors="coerce")
oop_df = oop_df.dropna(subset=["oop_share"])
oop_df = oop_df.sort_values(["Country Name", "year"])

oop_latest = (
    oop_df.groupby("Country Name", as_index=False)
    .last()
    .rename(columns={"Country Name": "Country"})
)

oop_latest["coverage_rate"] = 1 - oop_latest["oop_share"] / 100

oop_latest = oop_latest[["Country", "year", "oop_share", "coverage_rate"]]

print("Extracted latest coverage data:")
print(oop_latest.head(10))
print(f"\nTotal countries covered: {oop_latest['Country'].nunique()}")
```

```python
# Manual Match
name_map = {
    "Brunei": "Brunei Darussalam",
    "Czech Republic": "Czechia",
    "Egypt": "Egypt, Arab Rep.",
    "Hong Kong": "China",
    "Macau": "China",
    "Laos": "Lao PDR",
    "Russia": "Russian Federation",
    "Slovakia": "Slovak Republic",
    "South Korea": "Korea, Rep.",
    "Turkey": "Turkiye",
    "Vietnam": "Viet Nam",
    "Taiwan": "China",
    "Bermuda": "United Kingdom",
    "Guernsey": "United Kingdom",
    "Isle of Man": "United Kingdom",
    "Jersey": "United Kingdom",
    "French Guiana": "France",
    "Curacao": "Netherlands",
    "Puerto Rico": "United States",
    "Kosovo": "Serbia"
}

pharmacy_weighted["Countries"] = pharmacy_weighted["Countries"].replace(name_map)

import re, unicodedata
def clean_name(s):
    if pd.isna(s):
        return s
    s = re.sub(r'[\u2000-\u200b\u202f\u205f\u00a0]', ' ', s)
    s = unicodedata.normalize('NFKC', s.strip())
    return s

pharmacy_weighted["Countries"] = pharmacy_weighted["Countries"].apply(clean_name)
oop_latest["Country"] = oop_latest["Country"].apply(clean_name)

merged = pharmacy_weighted.merge(
    oop_latest[["Country", "coverage_rate"]],
    left_on="Countries",
    right_on="Country",
    how="left"
)
```

```python
#Compute per-capita and per-household health costs
```

```python
GLOBAL_BASELINE_HEALTH_EXP = 1200   # WHO global per-capita health expenditure (PPP USD)

merged["health_cost_per_capita"] = (
    GLOBAL_BASELINE_HEALTH_EXP *
    merged["weighted_pharmacy_index"] /
    merged["weighted_pharmacy_index"].mean()
)

# Out-of-pocket component
merged["oop_cost"] = merged["health_cost_per_capita"] * (1 - merged["coverage_rate"])


# Normalize for cross-country comparison (index = 100 avg)
merged["health_cost_index"] = (
    merged["oop_cost"] / merged["oop_cost"].mean() * 100
)
```

```python
# Assign frequency
freq_map_personal_care = {
    "Dry Cleaning": 4,          # Quarterly average  IBISWorld (2022): 35 times/year
    "Man's Haircut": 8,         # Every ~6 weeks  Statista Grooming Survey 2022
    "Woman's Haircut": 4,       # Every ~3 months  Statista Haircare Frequency 2022
    "Feminine products": 12,    # Monthly cycle  WHO/UNFPA hygiene standards
    "Toothpaste": 12,           # Monthly tube replacement  WHO Oral Health Survey 2019
    "Mouthwash": 6,             # Every 2 months  Mintel Oral Care 2023
    "Razor Blades": 6,          # Every 2 months  Statista Razor Usage 2022
    "Bath Soap": 12,            # Monthly bar/bottle  Euromonitor Hygiene Report 2023
    "Laundry Detergent": 12,    # Monthly household purchase  Euromonitor Cleaning 2022
    "Diapers": 4,               # Quarterly average per capita  UNICEF Diaper Market 2021
    "Shampoo": 12,              # Monthly  Nielsen Haircare Usage 2022
    "Nail Polish": 4,           # Quarterly  Mintel Beauty Report 2023
    "Deodorant": 12             # Monthly  Nielsen Global Deodorant Market 2022
}

canon_lookup = {k.lower(): v for k, v in freq_map_personal_care.items()}

def assign_freq(name):
    s = str(name).lower().strip()
    return canon_lookup.get(s, np.nan)

personal_care_df["freq_weight"] = personal_care_df["Final_Item"].apply(assign_freq)
```

```python
# Adjust frequency for different countries
income_adj = {
    "High income": 1.00,
    "Upper middle income": 0.85,
    "Lower middle income": 0.65,
    "Low income": 0.45
}
```

```python
# Calculation
```

```python
merged_personal["income_factor"] = merged_personal["IncomeGroup"].map(income_adj).fillna(1.0)
```

```python
merged_personal["freq_weight_adj"] = merged_personal["freq_weight"] * merged_personal["income_factor"]
```

```python
def annual_cost(df):
    df = df.dropna(subset=["Low_Price_usd", "freq_weight_adj"])
    if len(df) == 0:
        return np.nan
    df["annual_expense"] = df["Low_Price_usd"] * df["freq_weight_adj"]
    return df["annual_expense"].mean()
```

```python
personal_weighted = (
    merged_personal.groupby("Countries", as_index=False)
    .apply(lambda g: pd.Series({"personal_care_cost_usd": annual_cost(g)}),
           include_groups=False)
)
```
