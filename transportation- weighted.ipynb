```python
import mysql.connector

conn = mysql.connector.connect(
    host="studentdb.crsqkeh7g4o2.us-east-2.rds.amazonaws.com",
    user="living_wage",
    password="LivingWage123!",
    database="giq_data",
    port=3306,
    charset="utf8mb4",
    collation="utf8mb4_unicode_ci"
)

cursor = conn.cursor()
cursor.execute("SHOW TABLES;")
tables = cursor.fetchall()

for (table_name,) in tables:
    print(f"\nTable: {table_name}")
    
    # Get column names
    cursor.execute(f"DESCRIBE {table_name};")
    columns = [col[0] for col in cursor.fetchall()]
    
    # Build query to count NULLs
    null_checks = ", ".join(
        [f"SUM(CASE WHEN `{col}` IS NULL THEN 1 ELSE 0 END) AS `{col}_nulls`" for col in columns]
    )
    query = f"SELECT {null_checks} FROM `{table_name}`;"
    
    cursor.execute(query)
    null_counts = cursor.fetchone()
    
    # Print nicely
    for col, null_count in zip(columns, null_counts):
        print(f"{col}: {null_count} nulls")

cursor.close()
conn.close()
```

```python
import pandas as pd

df_t = df_transport.copy()
df_t['date'] = pd.to_datetime(df_t['date'], errors='coerce')
df_t['country']  = df_t['Countries'].str.strip()
df_t['currency'] = df_t['currency'].str.strip() if 'currency' in df_t.columns else 'NA'
df_t['fi_key']   = df_t['Final_Item'].str.strip().str.lower()

df_latest = (
    df_t.sort_values(['country','fi_key','date'], ascending=True, na_position='first')
        .groupby(['country', 'fi_key'], as_index=False)
        .tail(1)
)
```

```python
def to_monthly(row):
    lp  = float(row['Low_Price'])
    key = row['fi_key']
    if key == 'oil change':
        v = lp / 6
        ##per half year
        ##Filter Manufacturers Community
        #https://www.autocare.org/docs/default-source/communities-files/fmc/bulletins/94-1-automotive-oil-change-intervals-severe-vs-normal-driving     
    elif key == 'tires':
        v = lp / 72
        #per 3 year
        #National Highway Traffic Safety Administration
        #https://www.nhtsa.gov/vehicle-safety/tires
    elif key == 'gasoline':
        v = round(lp * (19840*7.2/1200), 3)
        ##https://www.eia.gov/todayinenergy/prices.php
        ##U.S. Energy Information Administration
        ##low price shows 0.7$,refer to 0.7/L. 
        ##Fuel Economy in Major Car Markets
        ##https://www.iea.org/reports/fuel-economy-in-major-car-markets
        ##7.2L/100km
        ##HOW MANY PEOPLE DOES ONE CAR AMOUNT TO? 2007
        ##https://ponce.sdsu.edu/cars_and_people.html
        ##19840km per year per car
        ##fomula would be lp*(19840*7.2/1200)

    elif key == 'car insurance':
        v = lp
        #refer to the low price of monthly car insurance cost 
        #Average Cost of Car Insurance in the US 2025
        #https://www.experian.com/blogs/ask-experian/average-cost-car-insurance
    elif key == 'car battery':
        v = lp / 36
        #per 3 year 
        ##Survey from 2025 Commercial Tire
        ##https://commercialtire.com/blog/how-often-to-replace-a-car-battery/
    elif key in ('subway', 'bus'):
        v = lp * 4   
        ##refer the low price of bus/suway weekly tickets
        ##https://www.mta.info/fares-tolls/subway-bus
        ##https://www.transitchicago.com/fares/
        ##https://www.sfgate.com/bayarea/article/bart-muni-increased-prices-starting-jan-1-20007254.php
        ##Therefore, for a monthly cost, multiply the price by 4.
    elif key == 'taxi':
        v = lp * 60
        ##refer to the low price of on taxi service
        #The Cost Per Mile For Driving, Taxiing, and Public Transport in America
        #https://www.self.inc/info/cost-per-mile-in-america/
    # elif key == 'car':
    #     v = lp
       
    else:
        v = pd.NA
    return v

df_latest['monthly_cost'] = df_latest.apply(to_monthly, axis=1)
```

```python
def to_bucket(key):
    if key in ('bus', 'subway'):
        return 'public_transport'
    elif key == 'taxi':
        return 'taxi'
    else:
        return 'car'
df_latest['bucket'] = df_latest['fi_key'].apply(to_bucket)
```

```python
per_item_monthly = (
    df_latest
    .pivot_table(index=['country','currency'], columns='fi_key',
                 values='monthly_cost', aggfunc='first')
    .reset_index()
)
per_item_monthly.columns.name = None

def bucket_agg(g):
    pt = g.loc[g['fi_key'].isin(['bus','subway']), 'monthly_cost'].mean(skipna=True)
    tx = g.loc[g['fi_key'].eq('taxi'), 'monthly_cost'].sum(skipna=True)
    car = g.loc[~g['fi_key'].isin(['bus','subway','taxi']), 'monthly_cost'].sum(skipna=True)
    return pd.Series({'public_transport': pt, 'taxi': tx, 'car': car})

per_bucket_monthly = (
    df_latest
    .groupby(['country','currency'], as_index=False)
    .apply(bucket_agg)
    .reset_index(drop=True)
)

print(per_bucket_monthly.head())


# per_bucket_monthly = (
#     df_latest
#     .groupby(['country','currency','bucket'], as_index=False)['monthly_cost']
#     .sum()
#     .pivot(index=['country','currency'], columns='bucket', values='monthly_cost')
#     .reset_index()
#     .fillna(0)
# )
```

```python
## consider the different weight of transportation categories in different contries
## ITF Transport Outlook(2023)
##https://www.oecd.org/content/dam/oecd/en/publications/reports/2023/05/itf-transport-outlook-2023_4466cd78/b6cc9ad5-en.pdf
region_car_share = {
    'Europe': 0.48,         # 48%
    'ENEA':   0.46,         # East & Northeast Asia
    'SEA':    0.52,         # 5052% 
    'SSWA':   0.24,
    'MENA':   0.27,
    'LAC':    0.33,         # 3233% 
    'SSA':    0.16,
    'TAP':    0.48,         # 4748% 
    'UCAN':   0.77,
}
```

```python
dfw = per_bucket_monthly.copy()
dfw['country_norm'] = dfw['country'].str.strip()
dfw['region'] = dfw['country_norm'].map(country_to_region)
dfw['car_share'] = dfw['region'].map(region_car_share).astype(float)
dfw['noncar_share'] = (1 - dfw['car_share']).round(2)
```

```python
dfw = dfw.dropna(subset=['car_share'])
```

```python
W_PT = 0.5  
W_TX = 0.5  

def noncar_cost_row(r):
    pt = r.get('public_transport', np.nan)
    tx = r.get('taxi', np.nan)
    if pd.notna(pt) and pd.notna(tx):
        return W_PT*pt + W_TX*tx
    elif pd.notna(pt):
        return pt
    elif pd.notna(tx):
        return tx
    else:
        return np.nan

dfw['noncar_cost'] = dfw.apply(noncar_cost_row, axis=1)
```

```python
# fomula Weighted = car_share * car_cost + (1 - car_share) * noncar_cost
dfw['weighted_monthly_cost'] = dfw['car_share']*dfw['car'] + dfw['noncar_share']*dfw['noncar_cost']

weighted_result = dfw[['country','region','currency','public_transport','taxi','car',
                       'car_share','noncar_share','noncar_cost','weighted_monthly_cost']].sort_values('country')

print(weighted_result.head(20))

print(weighted_result[weighted_result['country'].str.strip().str.lower()=='united states'])

weighted_result.to_csv('transport_weighted_monthly_by_country.csv', index=False, encoding='utf-8-sig')
```
